```python
!pip install bertviz
from transformers import GPT2Tokenizer, GPT2LMHeadModel

# Cargamos el tokenizador y el modelo preentrenado GPT-2
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
model = GPT2LMHeadModel.from_pretrained("gpt2")

# Definimos el texto de entrada
input_text = "In the world of artificial intelligence, models like GPT-2 have shown remarkable capabilities."

# Tokenizamos el texto de entrada para convertirlo en IDs comprensibles por el modelo
input_ids = tokenizer(input_text, return_tensors="pt").input_ids

# Generamos una secuencia de salida basada en el texto de entrada
output_ids = model.generate(input_ids, max_length=50, num_return_sequences=1)

# Decodificamos la secuencia de salida para convertirla de nuevo en texto legible
generated_text = tokenizer.decode(output_ids[0], skip_special_tokens=True)
print(generated_text)

# Utilizamos bertviz para visualizar las atenciones del modelo
from bertviz import head_view

# Obtenemos las atenciones del modelo
outputs = model(input_ids, output_attentions=True)
attentions = outputs.attentions

# Visualizamos las atenciones usando bertviz
head_view(attentions, tokenizer.convert_ids_to_tokens(input_ids[0]))

analisis="""La generación de texto con modelos como GPT-2 permite crear contenido de forma automatizada,
 ofreciendo nuevas posibilidades en la creación de textos, asistencias en redacción y más.
 La visualización de las atenciones ayuda a entender cómo el modelo procesa y genera el texto,
 brindando una herramienta educativa para explorar y mejorar estos modelos."""

print(analisis)
```
