*Hugging Face* es una librer√≠a que nos permite trabajar con una gran variedad de modelos basados en Transformers de una manera sencilla.

El modo m√°s simple de interactuar con los modelos de esta librer√≠a es utilizando el m√©todo ***pipeline()***. Esta funci√≥n se encargar√° autom√°ticamente de tokenizar tu texto y de hacer uso del modelo adecuado para resolver tu tarea. Por ejemplo, si quisieramos utilizar un clasificador de sentimientos, esto ser√≠a tan sencillo como:



```python
from transformers import pipeline
```

```python
# Creamos un clasificador para an√°lisis de sentimiento.
classifier = pipeline('sentiment-analysis')

# Y probamos qu√© tan bien funciona.
print(classifier('We are very happy to show you the Transformers library.'))
print(classifier('I hope this library is not as bad as programming everything from scratch.'))
```

```python
# Creamos un clasificador para an√°lisis de sentimiento con un modelo personalizado.
classifier = pipeline('sentiment-analysis', model='nlptown/bert-base-multilingual-uncased-sentiment')

# Probamos nu√©vamente el modelo en Espa√±ol.
print(classifier('¬°Est√° super bien esta librer√≠a!'))
print(classifier('Espero que esto no sea tan pesado como tener que programar todo desde cero.'))
print(classifier('Porque detestar√≠a tener que hacerlo.'))
```
